```python
def draw_barnsley_fern(t, num_points=50000, color="green", size=2):
    # Начальные координаты
    x, y = 0, 0
    
    # Коэффициенты преобразований для папоротника Барнсли
    # Каждое преобразование имеет вероятности и коэффициенты
    transforms = [
        (0.01, 0.00, 0.00, 0.16, 0.00, 0.00),
        (0.85, 0.04, -0.04, 0.85, 0.00, 1.60),
        (0.07, 0.20, 0.23, 0.22, 0.00, 1.60),
        (0.07, -0.20, 0.26, 0.24, 0.00, 0.44),
    ]
    
    # Масштабирование для отображения
    scale = 40
    offset_x = 0
    offset_y = -200
    
    # Рисование фрактала
    for i in range(num_points):
        print(f'{i}', end='\r')
        # Выбор преобразования на основе вероятностей
        r = random.random()
        
        if r <= 0.01:
            a, b, c, d, e, f = transforms[0]
        elif r <= 0.86:  # 0.01 + 0.85
            a, b, c, d, e, f = transforms[1]
        elif r <= 0.93:  # 0.01 + 0.85 + 0.07
            a, b, c, d, e, f = transforms[2]
        else:
            a, b, c, d, e, f = transforms[3]
        
        # Применение аффинного преобразования
        new_x = a * x + b * y + e
        new_y = c * x + d * y + f
        
        x, y = new_x, new_y
        
        # Преобразование координат для отображения
        plot_x = x * scale + offset_x
        plot_y = y * scale + offset_y
        
        # Рисование точки
        t.goto(plot_x, plot_y)
        t.dot(size, color)
```

```python
def pythagoras_tree(t, length, level):
    """Рекурсивное построение дерева Пифагора"""
    if level == 0:
        return
    
    # Рисуем ствол
    t.forward(length)
    
    # Сохраняем позицию и угол для правой ветви
    pos = t.position()
    angle = t.heading()
    
    # Правая ветвь (поворот на 45 градусов)
    t.right(45)
    pythagoras_tree(t, length * math.sqrt(2)/2, level - 1)
    
    # Возвращаемся к сохраненной позиции
    t.setposition(pos)
    t.setheading(angle)
    
    # Левая ветвь (поворот на -45 градусов)
    t.left(45)
    pythagoras_tree(t, length * math.sqrt(2)/2, level - 1)
    
    # Возвращаемся к началу ствола
    t.setposition(pos)
    t.setheading(angle)
```

```python
def koch_curve(t, length, depth):
    """Рекурсивное построение кривой Коха"""
    if depth == 0:
        t.forward(length)
        return
    
    length /= 3.0
    koch_curve(t, length, depth - 1)
    t.left(60)
    koch_curve(t, length, depth - 1)
    t.right(120)
    koch_curve(t, length, depth - 1)
    t.left(60)
    koch_curve(t, length, depth - 1)

def koch_snowflake(t, length, depth):
    """Снежинка Коха - три кривые Коха"""
    for _ in range(3):
        koch_curve(t, length, depth)
        t.right(120)
```

```python
t.reset()
t.speed(9)

t.penup()
t.goto(-150, 100)
t.pendown()

koch_snowflake(t, 300, 2)
```

```python
def sierpinski(t, length, depth):
    """Рекурсивное построение треугольника Серпинского"""
    if depth == 0:
        # Рисуем заполненный треугольник
        t.begin_fill()
        for _ in range(3):
            t.forward(length)
            t.left(120)
        t.end_fill()
    else:
        # Рекурсивно рисуем три меньших треугольника
        sierpinski(t, length/2, depth-1)
        t.forward(length/2)
        sierpinski(t, length/2, depth-1)
        t.backward(length/2)
        t.left(60)
        t.forward(length/2)
        t.right(60)
        sierpinski(t, length/2, depth-1)
        t.left(60)
        t.backward(length/2)
        t.right(60)
```

```python
t.reset()
t.speed(10)

t.color('blue', 'lightblue')

t.penup()
t.goto(-200, -150)
t.pendown()

sierpinski(t, 400, 5)
```
