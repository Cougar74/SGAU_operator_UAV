```python
def barnsley_fern(t, iterations):
    x, y = 0, 0  # Начальные координаты 
    for i in range(iterations):
        print(f'{i}', end='\r')
        r = random.random()  # Генерируем случайное число от 0 до 1
        # В зависимости от значения r выбираем одну из четырёх формул для новой точки
        if r < 0.01:
            x, y = 0, 0.16 * y                  # Стебель папоротника
        elif r < 0.86:
            x, y = 0.85 * x + 0.04 * y, -0.04 * x + 0.85 * y + 1.6   # Основная масса листьев
        elif r < 0.93:
            x, y = 0.2 * x - 0.26 * y, 0.23 * x + 0.22 * y + 1.6    # Левые листики
        else:
            x, y = -0.15 * x + 0.28 * y, 0.26 * x + 0.24 * y + 0.44 # Правые листики
        
        # Перемещаемся к новой точке 
        t.goto(x * 50, y * 50 - 200)
        t.dot(2, "yellow")  # Ставим точку диаметром 2 пикселя жёлтым цветом
```

```python
def pythagoras_tree(t, length, level):
    """Рекурсивное построение дерева Пифагора"""
    if level == 0:
        return
    
    # Рисуем ствол
    t.forward(length)
    
    # Сохраняем позицию и угол для правой ветви
    pos = t.position()
    angle = t.heading()
    
    # Правая ветвь (поворот на 45 градусов)
    t.right(45)
    pythagoras_tree(t, length * math.sqrt(2)/2, level - 1)
    
    # Возвращаемся к сохраненной позиции
    t.setposition(pos)
    t.setheading(angle)
    
    # Левая ветвь (поворот на -45 градусов)
    t.left(45)
    pythagoras_tree(t, length * math.sqrt(2)/2, level - 1)
    
    # Возвращаемся к началу ствола
    t.setposition(pos)
    t.setheading(angle)
```

```python
def koch_curve(t, length, depth):
    """Рекурсивное построение кривой Коха"""
    if depth == 0:
        t.forward(length)
        return
    
    length /= 3.0
    koch_curve(t, length, depth - 1)
    t.left(60)
    koch_curve(t, length, depth - 1)
    t.right(120)
    koch_curve(t, length, depth - 1)
    t.left(60)
    koch_curve(t, length, depth - 1)

def koch_snowflake(t, length, depth):
    """Снежинка Коха - три кривые Коха"""
    for _ in range(3):
        koch_curve(t, length, depth)
        t.right(120)
```

```python
t.reset()
t.speed(9)

t.penup()
t.goto(-150, 100)
t.pendown()

koch_snowflake(t, 300, 2)
```

```python
def sierpinski(t, length, depth):
    """Рекурсивное построение треугольника Серпинского"""
    if depth == 0:
        # Рисуем заполненный треугольник
        t.begin_fill()
        for _ in range(3):
            t.forward(length)
            t.left(120)
        t.end_fill()
    else:
        # Рекурсивно рисуем три меньших треугольника
        sierpinski(t, length/2, depth-1)
        t.forward(length/2)
        sierpinski(t, length/2, depth-1)
        t.backward(length/2)
        t.left(60)
        t.forward(length/2)
        t.right(60)
        sierpinski(t, length/2, depth-1)
        t.left(60)
        t.backward(length/2)
        t.right(60)
```

```python
t.reset()
t.speed(10)

t.color('blue', 'lightblue')

t.penup()
t.goto(-200, -150)
t.pendown()

sierpinski(t, 400, 5)
```
