```python
print(f'До: {os.listdir()}')

# Удаляет файлы и папки по пути
os.removedirs('./test')

print(f'После: {os.listdir()}')
```

```python
print(f'До: {os.listdir()}')

# Переименовывание файлов и папок
os.rename('./data_write.txt', './data_renamed_write.txt')

print(f'После: {os.listdir()}')
```

## `datetime`

```python
from datetime import datetime
```

```python
dt = datetime.today()

print(f'{dt}')
print(f'{dt.year} / {dt.month} / {dt.day}')
print(f'{dt.hour} : {dt.minute} : {dt.second}')
print(f'{dt.date()}')
print(f'{dt.time()}')
```

```python
dt = datetime.now()
print(f'{dt}')

# Количество секунд считается с 01.01.1970
print(f'Секунд: {dt.timestamp()}') # 1_764_778_736.88042
```

### Форматирование дат и времени

```python
now = datetime.now()

temp = now.strftime('%d.%m.%Y')
print(f'{type(now)} -- {type(temp)}')

print(now.strftime('%d.%m.%Y'))
print(now.strftime('%d/%m/%y'))
print(now.strftime('%d - %m - %Y\n'))

print(now.strftime('%H:%M:%S'))
print(now.strftime('%d.%m.%Y %H:%M\n'))

print(now.strftime("%A, %d %B %Y"))
print(now.strftime("%a, %d %b %Y"))
```

```python
date_str = "2025-12-31 23:59:59"

dt = datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')
print(f'{dt = }')
print(f'{type(dt) = }')
```
