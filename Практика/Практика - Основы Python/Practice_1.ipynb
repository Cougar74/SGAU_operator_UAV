# Типы данных

## Целые числа - `int`

```python
print(f'{2 + 3 = }')
print(f'{2 - 3 = }')
print(f'{2 * 3 = }')
print(f'{3 / 2 = }')
print(f'{3 ** 3 = }')
```

```python
print(f'{2 + 3 * 4 = }')
print(f'{(2 + 3) * 4 = }')
```

```python
print(f'{2 + 2 * 2 = }')
```

## Вещественные числа - `float`

```python
print(f'{0.1 + 0.1 = }')
print(f'{0.2 + 0.2 = }')
print(f'{2 * 0.1 = }')
print(f'{2 * 0.2 = }')
print(f'{0.4 / 0.2 = }')
```

```python
print(f'{0.2 + 0.1 = }')
print(f'{3 * 0.1 = }')

print(f'{round(0.2 + 0.1, 1) = }')
print(f'{round(3 * 0.1, 1) = }')
```

```python
print(f'{4 / 2 = }')
print(f'{1 + 2.0 = }')
print(f'{2 * 3.0 = }')
print(f'{3.0 ** 2 = }')
```

## Комплексные числа - `complex`

```python
comp = 3 + 4j
print(comp.real) # 3.0
print(comp.imag) # 4.0
```

```python
comp_1 = 3 + 4j
comp_2 = 2 + 5j
print(f'{comp_1 + comp_2 = }')
print(f'{comp_1 - comp_2 = }')
print(f'{comp_1 * comp_2 = }')
print(f'{comp_1 / comp_2 = }')
```

```python
# Изменение регистра символов в строках
name = "иванов иван иванович"
print(f'{name.title() = }')
```

```python
name = "Иванов Иван Иванович"
print(f'Переводим в верхний регистр: {name.upper()}')
print(f'Переводим в нижний регистр: {name.lower()}')
print(f'Меняем регистры на противоположные: {name.swapcase()}')
print(f"{'иванов иван иванович'.capitalize() = }")
```

```python
# иванов иван иванович -> иВАНОВ иВАН иВАНОВИЧ
name = "иванов иван иванович"

'''
name = name.title()
name = name.swapcase()
print(name)
'''

print(name.title().swapcase())
print('Иванов Иван Иванович'.swapcase())
```

```python
# Поиск и замена
example_text = "hello world hello"

print(f"{example_text.find('hello') = }")
print(f"{example_text.find('hello', 1) = }")
print(f"{example_text.rfind('hello') = }")
print(f"{example_text.find('python') = }")
```

```python
example_text = "hello world hello"

print(f"{example_text.replace('hello', 'hi') = }")
print(f"{example_text.replace('hello', 'hi', 1) = }")
```

```python
# Проверка содержимого
print(f"{'hello'.isalpha() = }")
print(f"{'123'.isdigit() = }")
print(f"{'hello123'.isalnum() = }")
print(f"{'   '.isspace() = }")
print(f"{'Hello World'.istitle() = }")
print(f"{'HELLO'.isupper() = }")
print(f"{'hello'.islower() = }")

print(f"{'hello'.startswith('he') = }")
print(f"{'hello'.startswith('el') = }")
print(f"{'world'.endswith('rld') = }")
```

```python
# Разделение текста
example_text = "apple,banana,orange"
fruits = example_text.split(',')
print(f"{fruits = }\n")

print(f"{'Hello World Python'.split() = }")
print(f"{'one two three four'.split(' ', 2) = }\n")
```

## f-string

```python
a, b = 10, 3

print(f"Сумма: {a + b}")
print(f"Разность: {a - b}")
print(f"Сравнение: {a > b}")
print(f"Вызов метода: {'hello'.upper()}")
```

```python
def square(x):
  return x * x

print(f"Квадрат числа: {square(5)}")
print(f"Квадрат числа: {square(32)}")
```

```python
# Целые числа
number = 42

print(f"Обычное: {number}")
print(f"С заполнением: {number:05d}")
print(f"С разделителем: {10**16} или {10**16:,}")
print(f"В двоичной: {number:b}")
print(f"В шестнадцатеричной: {number:X}")
```

```python
# Дробные числа
pi = 3.14159265

print(f"Обычное: {pi}")
print(f"2 знака: {round(pi, 2)} или {pi:.2f}")
print(f"Процент: {0.75:.1%}")
print(f"Научная нотация: {10**16:.2e}")
```

```python
# Выравнивание текста
text = "Python"
number = 42

print(f'|{text:<15}|')  # |Python  | - по левому краю
print(f'|{text:>15}|')  # |  Python| - по правому краю
print(f'|{text:^15}|')  # | Python | - по центру
print(f'|{number:>15d}|') # |    42| - числа тоже можно выравнивать
```

```python
# Выравнивание текста c заполением
text = "Python"

print(f'|{text:*<15}|')
print(f'|{text:->15}|')
print(f'|{text:=^15}|')
```

```python
example_text = ' python '
print(f'Исходный текст: _{example_text}_')
print(f'Текст без пробелов справа: _{example_text.rstrip()}_')
print(f'Текст без пробелов слева: _{example_text.lstrip()}_')
print(f'Текст без пробелов: _{example_text.strip()}_\n')

example_text = '***hello***'
print(f'Исходный текст: {example_text}')
print(f"{example_text.strip('*') = }")
print(f"{example_text.lstrip('*') = }")
print(f"{example_text.rstrip('*') = }")
```

```python
# Срезы - slice
example_text = 'Hello World'
print(f"{' '.join(f'{v:<4}' for v in example_text)}")
print(f"{' '.join(f'{i:<4}' for i in range(len(example_text)))}\n")

print(f'{example_text[0:5] = }')
print(f'{example_text[6:] = }')
print(f'{example_text[5:] = }')
print(f'{example_text[:5] = }')
print(f'{example_text[:] = }')
print(f'{example_text[::2] = }')
print(f'{example_text[::-1] = }')
```
