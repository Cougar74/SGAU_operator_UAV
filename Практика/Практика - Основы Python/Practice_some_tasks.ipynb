# Задача #1  
Есть лестница. У лестницы некоторое `N` ступенек. Мы можем шагать на 1 или на 2 ступенки за один ход.  
Как много способов подняться по лестнице?  

**Пример**
n = 2 -> Ответ: 2  
1. 1 step + 1 step
2. 2 steps

```python
N = 3 # 3 решения
```

```python
def climbStairs(n):
    res = 0
    
    if n:
        if n > 1:
            res += climbStairs(n - 1)
            res += climbStairs(n - 2)
        else:
            res += climbStairs(n - 1)
    else:
        res = 1
    
    return res

def climbStairsOpt(n):
    if n == 0 or n == 1:
        return 1

    return climbStairsOpt(n - 1) + climbStairsOpt(n - 2)
```

```python
climbStairs(2), \
climbStairsOpt(2)
```

```python
climbStairs(4), \
climbStairsOpt(4)
```

#  НОД и НОК

НОД или gcd: НОД(18, 45)  
18 = 9 * 2 = 3 * 3 * 2  
45 = 9 * 5 = 3 * 3 * 5  
НОД(18, 45) = 9

НОК или lcm: НОК(18, 45)  
НОК(18, 45) = 90

```python
def multiplier(a: int) -> list[int]:
    # Выделяет простые множители числа
    i = 2
    temp = a
    result = []
    
    while temp > 1:
        if i > temp:
            result.append(temp)
            break
        
        if not temp % i:
            result.append(i)
            temp //= i
        else:
            i += 1
            
    return result

def gcd(a: int, b: int) -> int:
    # НОД
    multipliers_a = multiplier(a)
    multipliers_b = multiplier(b)
    
    m_union = []
    for item in multipliers_a:
        if item in multipliers_b:
            m_union.append(item)
            multipliers_b.remove(item)
    
    result = 1
    for item in m_union:
        result *= item
        
    return result

def gcd_opt(a: int, b: int) -> int:
    while b:
        # print(f'Before: {a} -- {b} ... {a % b = }')
        a, b = b, a % b
        # print(f'After: {a} -- {b}\n')
    return abs(a)

def lcm(a: int, b: int) -> int:
    # НОК
    temp = max(a, b)
    
    print(f'Start: {temp = }')
    while temp % a:
        print(f'{temp}: {temp % a = }')
        temp += max(a, b)
    
    return temp

def gcd_to_lcm(a: int, b: int) -> int:
    if a == 0 or b == 0:
        return 0
    return abs(a * b) // gcd_opt(a, b)
```

```python
m_a = [2, 3, 3]
m_b = [3, 3, 5]

m_union = []
for item in m_a:
    if item in m_b:
        m_union.append(item)
        m_b.remove(item)
```

Сложность алгоритма  
multiplier -> O(n) x 2 -> O(2 * n) -> O(n)  
gcd -> O(n) * O(n) -> **$O(n^2)$**

```python
result = []

def letterBacktrack(digits, combination, index):
    print(f'#{index}: {combination} -- {result}')
    
    if index == len(digits):
        result.append(combination[:])
        return
        
    for letter in letters[digits[index]]:
        letterBacktrack(digits, combination + letter, index + 1)

def letterCombinations(digits: str):
    letterBacktrack(digits, '', 0)
```
