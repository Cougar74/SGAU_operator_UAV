## Классы

```python
class Dog():
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def sit(self):
        print(f'{self.name} сел!')
    
    def bark(self):
        print('Гав-гав!')
    
    def roll_over(self):
        print(f'{self.name} сделала перека!')
```

```python
my_dog = Dog('Банди', age=4)

print(f'{type(my_dog) = }')
```
